```python
import pandas as pd
```

```python
file_path = '/content/House in Mumbai - 810+ House for Sale in Mumbai.xlsx'
xls = pd.ExcelFile(file_path)

xls.sheet_names
```

```python
df = pd.read_excel(file_path, sheet_name='Sheet1')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
df
```

```python
def clean_rate_sqft(rate):
    if isinstance(rate, str):
        rate = rate.replace('₹', '').replace(',', '').split('/')[0]
    return pd.to_numeric(rate, errors='coerce')

df['Rate_SqFt'] = df['Rate_SqFt'].apply(clean_rate_sqft)
```

```python
def converter(x):
    if isinstance(x, str):
      x = x.replace('₹', '').replace(',', '').strip()
      if 'Lac' in x:
          return f"{(float(x.strip('Lac'))*1):,.1f}"
      elif 'Crore' in x:
          return f"{(float(x.strip('Crore'))*100):,.1f}"
    else:
      return x

df['Price_Lakh'] = df['Price'].apply(converter)
df["Price_Lakh"] = df["Price_Lakh"].str.replace(',','')
df = df.drop("Price", axis=1)
```

```python
def clean_plot_area(price):
  if isinstance(price, str):
    price = price.replace('sqft', '').replace(',', '').split('/')[0]
  return pd.to_numeric(price, errors='coerce')

df['Plot_area'] = df['Plot_area'].apply(clean_plot_area)
```

```python
import re

def clean_bathroom(bathroom):
    if isinstance(bathroom, str):

        match = re.search(r'\d+', bathroom)
        if match:
            return int(match.group())
    return pd.to_numeric(bathroom, errors='coerce')

df['Bathroom'] = df['Bathroom'].apply(clean_bathroom).astype('Int64')
```

```python
df['Bedroom'] = df['Bedroom'].str.replace('BHK', '').str.strip()
```

```python
def extract_location(address):
    if isinstance(address, str) and 'in' in address:
        return address.split('in')[-1].strip()
    return address

df['Location'] = df['Location'].apply(extract_location)
```

```python
df['Availability'] = df['Availability'].str.strip()
```

```python
df
```

```python
def extract_region(location):
  if isinstance(location, str):
    parts = location.split(',')
    if len(parts) > 1:
        return parts[1].strip()
    return None

df['Region'] = df['Location'].apply(extract_region)
```

```python
def alter_property_name(name):
  if isinstance(name, str):
    parts = name.split(',')
    if len(parts) > 1:
      return parts[0].strip()
    return None

df['Property_name'] = df['Property_name'].apply(alter_property_name)
df['Property_name']
```

```python
df
```

```python
df = df.drop(columns=["Field11"])
df.isna().sum()
```

```python
df = df.dropna(subset=['Property_name'])
df['Price_Lakh'] = pd.to_numeric(df['Price_Lakh'], errors='coerce')
df['Price_Lakh'] = df['Price_Lakh'].fillna(df['Price_Lakh'].median())
df.isna().sum()
```

```python

df['Rate_SqFt'] = df['Rate_SqFt'].fillna(df['Rate_SqFt'].mean())
df['Plot_area'] = df['Plot_area'].fillna(df['Plot_area'].mean())
```

```python
# Fill missing categorical columns with the most frequent value
df['Bedroom'] = df['Bedroom'].fillna(df['Bedroom'].mode()[0])
df['Bathroom'] = df['Bathroom'].fillna(df['Bathroom'].mode()[0])
df['Availability'] = df['Availability'].fillna(df['Availability'].mode()[0])
df['Region'] = df['Region'].fillna(df['Region'].mode()[0])
```

```python
df = df[['Property_name','Location','Region','Availability','Plot_area','Rate_SqFt','Bedroom','Bathroom','Price_Lakh']]
df
```

```python
mumbai_property = df.to_csv('Mumbai_Property.csv', index=False)
```
